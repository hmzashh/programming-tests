# **Univariate analysis**

# **on categorical data**

```python
cat_data=df.select_dtypes(include='object')
cat_data
cat_data1=['Ship Mode', 'Segment', 'Country', 'State', 'Region', 'Category', 'Sub-Category']
```

```python

for i in cat_data1:
  plt.figure(figsize=(10, 5))
  sns.countplot(x=cat_data[i])
  plt.xlabel(f'{i}')
  plt.ylabel('Count')
  plt.title(f'Distribution of {i}')
  plt.tight_layout()
  plt.xticks(rotation=45)
  plt.show()
```


# **on numerical data :**

*   The graph shows the distribution of the variable Sales, which is very skewed to the right.

*   The graph shows values ​​that are very far from the rest of the data (outliers) using a box plot.




```python
num_data=df.select_dtypes(include='number')
num_data.head(1)
```

```python
sns.histplot(df['Sales'], kde=True)
plt.title('Distribution of Sales')
plt.show()
```

```python
sns.boxplot(x=df['Sales'])
plt.title('Sales - Boxplot')
plt.show()
```

# **Bivariate analysis**

```python
df.head(1)
```

below shapes show total sales for each category





```python
cat_data1 = ['Ship Mode', 'Segment', 'Country', 'State', 'Region', 'Category', 'Sub-Category']
for col in cat_data1:
    sns.barplot(data=df, x=col, y='Sales', estimator=sum)
    plt.xlabel(f'{col}')
    plt.ylabel('Total Sales')
    plt.title(f'Total Sales Distribution by {col}')
    plt.tight_layout()
    plt.xticks(rotation=45)
    plt.show()
```

# **Multivariate**

The following figure shows the total sales for each sector, in addition to detailing each sector by color.


```python
sns.barplot(data=df, x='Segment', y='Sales', hue='Category', estimator=sum)
plt.title('Total Sales by Segment and Category')

plt.show()
```

The following figure shows the number of sales orders for each region, in addition to the sectors that each region orders and their frequency.


```python
sns.countplot(data=df, x='Region', hue='Segment')
plt.title('Segment Distribution across Regions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
