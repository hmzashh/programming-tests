```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/clv_data.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().mean() * 100
```

# **Handling missing value:**


*   filling age using median instead of mean because the distribution not normal
*   filling days_on_platform using median instead of mean because the distribution not normal




```python
df['days_on_platform'].skew()
```

```python
sns.histplot(df['days_on_platform'],kde=True)
plt.show()
```

```python
#handling missing value
df_new=df.copy()
df_new['age']=df_new['age'].fillna(df_new['age'].median())
df_new['days_on_platform']=df_new['days_on_platform'].fillna(df_new['days_on_platform'].median())
```

```python
df_new.isnull().mean() * 100
```

# **Handling Outliers**


*   income + day on platform + prurchases ,all of them are not normal dist and they have outliers that we will handle.



```python
num_col=['age', 'income', 'days_on_platform']
for col in num_col:
  sns.histplot(df_new[col],kde=True)
  plt.show()
```

```python
num_col=['age', 'income', 'days_on_platform']
for col in num_col:
  Q1=df_new[col].quantile(0.25)
  Q3=df_new[col].quantile(0.75)
  IQR=Q3-Q1
  Lower=Q1-1.5*IQR
  Upper=Q3+1.5*IQR
  df_new=df_new[(df_new[col]>=Lower)&(df_new[col]<=Upper)]
```

```python
num_col=['age', 'income', 'days_on_platform']
for col in num_col:
  sns.boxplot(df_new,x=col)
  plt.show()
```

# **Encoding**


*   encoding gender
*   encoding city



```python

categorical_cols = ['gender', 'city']

ohe = OneHotEncoder(sparse_output=False, drop=None)

encoded_array = ohe.fit_transform(df_new[categorical_cols])

encoded_df = pd.DataFrame(encoded_array, columns=ohe.get_feature_names_out(categorical_cols))

df_encoded = pd.concat([df_new.drop(columns=categorical_cols), encoded_df], axis=1)
df_encoded.head(15)
```
