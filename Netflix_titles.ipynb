```python
num_df=df.select_dtypes(include='number')
for num in num_df.columns:
  sns.histplot(df[num])
  plt.show()
```

```python
cat_df1=['type', 'country', 'date_added','rating', 'duration', 'listed_in',]
for i in cat_df1:
  print(i)
  print(df[i].value_counts())
  print('------------------------------------------')
```

```python
for col in cat_df1:

  top=df[col].value_counts().nlargest(12).index
  topdf=df[df[col].isin(top)]

  plt.figure(figsize=(10,5))
  sns.countplot(topdf,x=col,order=top)
  plt.xticks(rotation=45)
  plt.show()


```

**BIVARIATE ANALYSIS**

```python
sns.countplot(topdf,x='type',hue='rating')
plt.xticks(rotation=45)
plt.show()
```

The following figure shows that the TV-show type is classified as for children, while the movie type is for all interests, comedy, documentary, etc.


```python
sns.countplot(topdf,x='type',hue='listed_in')
plt.xticks(rotation=45)
plt.show()
```

The figure (country vs type) shows the interest of countries in the type they follow. For example, in the United States, the audience tends to follow movies more than the audience in Japan, which tends to follow TV shows more.
