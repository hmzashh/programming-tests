# **Univariate analysis**

the following histogram for parental level of education that show each level and count of people that studing the level


```python
#parental level of education
sns.countplot(df['parental level of education'])
plt.xlabel('count')
plt.show()
```

the following bar plot view gender type in dataset and count of them

```python
#bar plot
sns.countplot(x=df['gender'])
plt.xlabel('gender')
plt.ylabel('count')
plt.show()
```

the following bar plot show the count of race/ethnicity

```python
#bar
sns.countplot(x=df['race/ethnicity'])
plt.title('count')
plt.xlabel('race')
plt.ylabel('count')
plt.show()
```

the following pie chart view the percentage of lunch type for student

```python
#pie
labels=['standerd','free']
sizes=df['lunch'].value_counts()
colors=['green','red']
explode=(0,0.1)
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',explode=explode)
plt.title('pie chart')
plt.show()
```

```python
sns.boxplot(df,x='parental level of education',y='math score')
plt.title('math score by parental level of education')
plt.xlabel('parental level of education')
plt.xticks(rotation=45)

plt.ylabel('math score')
plt.show()
```

all student among all levels tend to eat standered lunch

```python
sns.countplot(df,x='parental level of education',hue='lunch')
plt.title('lunch by parental level of education')
plt.xlabel('parent level')
plt.xticks(rotation=45)

plt.ylabel('lunch')
plt.show()
```
